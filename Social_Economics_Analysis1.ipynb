## Melting

```python
years = [str(i) for i in range(1960,2017)]
```

Melting the Population Dataset

```python
population.columns
```

```python
population.shape
```

```python
population = pd.melt(population,
             id_vars =['Country Name','Country Code'],
             value_vars = years,
             var_name ='Year',
             value_name = 'population')

population.head()
```

```python
population.shape
```

Melting the Fertility Dataset

```python
fertility.columns
```

```python
fertility.shape
```

```python
fertility = pd.melt(fertility,
             id_vars =['Country Name','Country Code'],
             value_vars = years,
             var_name ='Year',
             value_name = 'fertility')
fertility.head()
```

```python
fertility.shape
```

Melting the Life Expectancy Dataset

```python
life_expectancy.columns
```

```python
life_expectancy = pd.melt(life_expectancy,
             id_vars =['Country Name','Country Code'],
             value_vars = years,
             var_name ='Year',
             value_name = 'life_expectancy')
life_expectancy.head()
```

Convert Year to Integer

```python
population['Year'] = population['Year'].astype(int)
fertility['Year'] = fertility['Year'].astype(int)
life_expectancy['Year'] = life_expectancy['Year'].astype(int)
```

## Merging

```python
# Merge with Region info from country metadata
df = pd.merge(country, population, on='Country Code')
df.head()
```

```python
# Merge with Fertility
df = pd.merge(df, fertility,
              on=['Country Name', 'Country Code', 'Year'],how='inner')
df.head()
```

```python
# Merge with Life Expectancy
df = pd.merge(df, life_expectancy,
              on=['Country Name', 'Country Code', 'Year'])
df.head()
```

```python
df.shape
```

```python
# reorder
df = df[['Country Name', 'Country Code', 'Region', 'IncomeGroup','Year',
         'population', 'fertility', 'life_expectancy']]

df.head()
```

# EDA

###  Line charts showing trends in life expectancy, fertility rate and population

```python
# Prepare trend data by year
life_expectancy_trend = df.groupby('Year')['life_expectancy'].mean().reset_index()
fertility_trend = df.groupby('Year')['fertility'].mean().reset_index()
population_trend = df.groupby('Year')['population'].mean().reset_index()
```

```python
life_expectancy_trend
```

```python
fertility_trend
```

```python
population_trend
```

```python
# Set figure size
plt.figure(figsize=(12, 4))

# Create the subplot for life expectancy
plt.subplot(1, 3, 1)
plt.plot(life_expectancy_trend['Year'], life_expectancy_trend['life_expectancy'])
plt.title('Global Average Life Expectancy Over Years')
plt.ylabel('Life Expectancy (Years)')


# Create the subplot for fertility rate
plt.subplot(1, 3, 2)
plt.plot(fertility_trend['Year'], fertility_trend['fertility'], color='orange')
plt.title('Global Average Fertility Rate Over Years')
plt.ylabel('Fertility Rate (Births per Woman)')


# Create the subplot for Population
plt.subplot(1, 3, 3)
plt.plot(population_trend['Year'], population_trend['population'], color='green')
plt.title('Global Total Population Over Years')
plt.ylabel('Population')
plt.xlabel('Year')


plt.tight_layout()
plt.show()
```

### Country wise Comparison

```python
# Choose countries
countries_to_compare = ['India', 'China', 'United States']

# Create a dictionary to store data for each country
country_data = {}

# Filter Data for Each Country
for country in countries_to_compare:
    country_data[country] = df[df['Country Name'] == country]
```

```python
country_data
```

```python
# Set figure size
plt.figure(figsize=(12, 4))

# Create the subplot for Life Expectancy
plt.subplot(1, 3, 1)
for country in countries_to_compare:
    plt.plot(country_data[country]['Year'], country_data[country]['life_expectancy'], label=country)
plt.title('Life Expectancy')
plt.ylabel('Age')
plt.legend()

# Create the subplot for Fertility Rate
plt.subplot(1, 3, 2)
for country in countries_to_compare:
    plt.plot(country_data[country]['Year'], country_data[country]['fertility'], label=country)
plt.title('Fertility Rate')
plt.ylabel('Births per Woman')

# Create the subplot for Population
plt.subplot(1, 3, 3)
for country in countries_to_compare:
    plt.plot(country_data[country]['Year'], country_data[country]['population'], label=country)
plt.title('Population')
plt.ylabel('Population')


# Adjust layout
plt.suptitle('Comparison of Countries', fontsize=16)
plt.tight_layout()
plt.show()
```
